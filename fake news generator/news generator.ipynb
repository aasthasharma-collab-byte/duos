```python
import random

subjects = ["Government", "Scientists", "Students", "Police"]

verbs = ["discover", "ban", "launch", "expose"]
objects = ["secret technology", "new AI", "alien signal", "ancient city"]

headline = f"{random.choice(subjects)} {random.choice(verbs)} {random.choice(objects)}"
print(headline)
```

```python
import random

score = random.randint(50,100)

print(f"Fake Score : {score}%")

if score > 80:
    print("Highly Clickbait")
elif score >60:
    print("Moderately Fake")
else:
    print("Looks believable")
```

```python
from textblob import TextBlob

headline = input("Enter headline : ")

analysis = TextBlob(headline)

print("Polarity :",analysis.sentiment.polarity)

if analysis.sentiment.polarity>0:
    print("Positive")
elif analysis.sentiment.polarity<0:
    print("Negative")
else:
    print("Neutral")
```

```python
headline = "Scientists discover alien technology"

keywords = ["discover","alien"]

for word in headline.split():
    if word.lower() in keywords:
        print("*",word)
```

```python
import sqlite3

conn = sqlite3.connect("headlines.db")
cur = conn.cursor()

cur.execute("""
CREATE TABLE IF NOT EXISTS headlines(
id INTEGER PRIMARY KEY,
headline TEXT
)
""")

headline="Scientists discover alien technology"

cur.execute("INSERT INTO headlines(headline) VALUES(?)",(headline,))

conn.commit()
conn.close()
```
